# Tipos de Dados

Python não é fortemente tipado, ou seja, ele define o tipo baseado no valor atribuído.

Utilizando comando **"type()"** é possivel extrair o tipo do dado.

```python
#Tipo numérico
idade = 13 #tipo int(integer/inteiro)
print(type(idade))
idade = ['1','2','3']

print(idade)
print(type(idade))
```

```python
#Tipo numérico
altura = 1.85 #tipo float(flutuante)
print(type(altura))
```

```python
#Tipo booleano
fumante = False
print(type(fumante))
```

```python
#Tipo string
nome_completo = 'Edson Arantes do Nascimento o Pelé'
print(type(nome_completo))
```

# Operações aritméticas

Operações básicas:
* '+': soma
* '-': subtração
* '*': multiplicação
* '/': divisão
* '**': exponenciação
* '%': módulo -> resto divisão

```python
#Soma
saldo = 10 + 3
print(saldo)

#subtração
conta_cc = 15 -7
print(conta_cc)

#multiplicação
soma_notas = 10 * 5
print(soma_notas)

#divisao
media = soma_notas/5
print(media)

#módulo
resto = 10%3
print(resto)

#exponenciação
log = 10**2
print(log)
```

# Listas

Method	Description
* append()	Adiciona um elemento na lista
* clear()	Remove todos os elementos da lista
* copy()	Retorna uma cópia da lista
* count()	Retorna o numero de elementos
* extend()	Add the elements of a list (or any iterable), to the end of the current list
* index()	Retorna o index do primeiro item e seu valor
* insert()	Adicion um elemento em uma posição especificad
* pop()	Remove um elemento em uma posição especoficada
* remove()	Remove um ele
* reverse()	Reverses the order of the list
* sort()	Sorts the list



```python
nome = 'Julio'
#       [   0   ,   1    ,   2   ]
nomes = ['Julio', 'Edson', 'Pelé'] #definir uma lista utiliza-se o colchetes
print(type(nome))
print(nome)
print(type(nomes))
print(nomes) #Apresenta todos os itens da lista
print(nomes[0] + ' ' + nomes[2]) #Coleta dp item da lista
nomes[2] = 10
print(nomes)
```

```python
notas = [] #Lista vazia
print(notas)
notas.append(9.9) #Adicionar um item na ultima posição da lista
print(notas)
notas.append(10)
print(notas)
```

```python
#Utilizando o método .append() crie uma lista contendo os nomes dos alunos da
#fileira da frente. Para inserir os nomes use o comando input()

nomes_alunos_fileira = [] #Criei a lista
#Aluno 1
nome_aluno = input('Digite o nome do aluno: ') #Recebo o nome vindo do usuário
nomes_alunos_fileira.append(nome_aluno) #Adicionei o nome do aluno
print(nomes_alunos_fileira) #Aprsentei os nomes
#Aluno 2
nome_aluno = input('Digite o nome do aluno: ') #Recebo o nome vindo do usuário
nomes_alunos_fileira.append(nome_aluno) #Adicionei o nome do aluno
print(nomes_alunos_fileira) #Aprsentei os nomes
#Aluno 3
nome_aluno = input('Digite o nome do aluno: ') #Recebo o nome vindo do usuário
nomes_alunos_fileira.append(nome_aluno) #Adicionei o nome do aluno
print(nomes_alunos_fileira) #Aprsentei os nomes
```

```python
desenhos = ['Mundo de Bob', 'Dragon Ball Z', 'Meninas Super-Poderosas', 'TinTin']
print(desenhos)
desenhos.remove('Mundo de Bob') #Deleta um item especificado
print(desenhos)
item_removido = desenhos.pop(1) #Deleta um item especificado pelo index da lista
print(item_removido)
print(desenhos)
```

```python
desenhos_90 = ['Mundo de Bob', 'Dragon Ball Z', 'Meninas Super-Poderosas', 'TinTin']
desenhos_CN = ['Charlie Brown', 'KND']

desenhos_totais = desenhos_CN.copy() #Copiei a lista desenhos_CN para uma nova lista
desenhos_totais.extend(desenhos_90)#Extendi a minha lista criada com desenhos_90
#desenhos_totais = desenhos_90 + desenhos_CN
print(desenhos_totais)
```
